```python
import pandas as pd

data = {
    'StudentID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Maths': [85, 90, 75, 80, 95],
    'Science': [88, 92, 80, 85, 90],
    'English': [90, 85, 78, 82, 88]
}

df = pd.DataFrame(data)
df
```

```python
def add_1(x):
    return x + 2
```

```python
# apply on single column
df['Maths'] = df['Maths'].apply(add_1)
df
```

```python
#  apply function on multiple columns
df[['Maths','Science']] = df[['Maths','Science']].apply(add_1)
df
```

```python
num_df = df[['StudentID', 'Maths','Science','English']]
num_df
```

```python
# apply function on whole dataframe
num_df = num_df.apply(add_1)
num_df
```

```python
# pass whole row to the function by applying axis=1
def add_marks(row):
    return row['Maths'] + row['Science'] + row['English']
```

```python
df['Total'] = num_df.apply(add_marks, axis=1)
df
```

```python
# pass selected colums in the row to the function by applying axis=1 and args
def calc_perc(row, col1, col2, col3):
    
    perc = ((row[col1] + row[col2] + row[col3]) / 300) * 100
    
    if perc > 90:
        Grade = 'A'
    elif perc >= 80 and perc < 90:
        Grade = 'B'
    elif perc >=70 and perc < 80:
        Grade = 'C'
    else:
        Grade = 'D'

    return Grade
        
        
```

```python
df['Grade'] = df.apply(calc_perc, args=('Maths', 'Science', 'English'), axis=1)
df
```

# lambda

```python
data = {
    'StudentID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Maths': [85, 90, 75, 80, 95],
    'Science': [88, 92, 80, 85, 90],
    'English': [90, 85, 78, 82, 88]
}

df = pd.DataFrame(data)
df
```

```python
df['Maths'] = df['Maths'].apply(lambda x:x+1)
df
```

```python
df[['Maths', 'Science']] = df[['Maths', 'Science']].apply(lambda x: x+1)
df
```

```python
df['Total'] = df.apply(lambda row, x,y,z: row[x] + row[y] + row[z], args=('Maths', 'Science', 'English'), axis=1)
df
```

```python
df['Name'] = df['Name'].str.upper()
df
```
