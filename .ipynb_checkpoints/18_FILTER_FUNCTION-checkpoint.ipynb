```python
import pandas as pd

data = {
    'StudentID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Maths': [85, 90, 75, 80, 95],
    'Science': [88, 92, 80, 85, 90],
    'English': [90, 85, 78, 82, 88]
}

df = pd.DataFrame(data)
df
```

```python
# select columns
df.filter(items=['Maths','Science'])
```

```python
# match on pattern
df.filter(like='en',axis=1)
```

```python
df.filter(regex='[DM]', axis=1)
```
