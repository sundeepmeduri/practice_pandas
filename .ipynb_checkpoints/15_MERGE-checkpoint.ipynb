```python
import pandas as pd

# Sample DataFrame 1
data1 = {'key': [1, 2, 3, 4],
         'value1': ['A', 'B', 'C', 'D']}
df1 = pd.DataFrame(data1)
df1 = df1.set_index('key')
# Sample DataFrame 2
data2 = {'key': [2, 3, 4, 5],
         'value2': ['E', 'F', 'G', 'H']}
df2 = pd.DataFrame(data2)
df2 = df2.set_index('key')
```

```python
df1
```

```python
df2
```

```python
# Inner join
inner_df = pd.merge(df1, df2, on='key', how='inner')
inner_df
```

```python
left_df = pd.merge(df1, df2, on='key', how='left')
left_df
```

```python
right_df = pd.merge(df1, df2, on='key', how='right')
right_df
```

```python
outer_df = pd.merge(df1, df2, on='key', how='outer')
outer_df
```
