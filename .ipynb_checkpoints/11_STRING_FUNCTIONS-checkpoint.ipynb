```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob ', 'Charlie ', 'David', 'Eve'],
    'City': ['New York  ', 'Los Angeles', 'Chicago', 'Houston  ', 'Phoenix'],
    'Country': ['USA', 'USA', 'USA', '  USA', 'USA   ']
}

df = pd.DataFrame(data)
df
```

```python
df = df.map(str.upper)
df
```

```python
df = df.map(str.title)
df
```

```python
# concat strings
df['concat'] = df['City'].str.strip() + '_' + df['Country'].str.strip()
df
```

```python
# concats columns to new dataframe
df2 = pd.concat([df['Name'], df['City']], axis=1)
df2
```

```python
df
df['concat'] = df['concat'].str.replace('_','-')
df
```

```python
df[['new_city', 'new_country']] = df['concat'].str.split('-',n=1,expand=True)
df
```
